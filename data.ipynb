```python
import requests
import json
import base64
import sqlite3
```

```python
url="http://192.168.6.210:8064/pipl/restworld/getAllCollegeStudents"
```

```python
response=requests.get(url)
```

```python
response
```

```python
if response.status_code == 200:
    data = response.json()  # Parsed JSON data
else:
    print("Failed to fetch data:", response.status_code)
```

```python
conn=sqlite3.connect('pimeet.db')
cursor=conn.cursor()
```

```python
for student in data:
    name = student['name']
    email = student['email']
    contact = student['contact']
    profile_pic_type = student['studentProfilePicType']
    
    # Decode profile pic if exists
    if student['studentProfilePic'].strip():
        try:
            profile_pic_blob = base64.b64decode(student['studentProfilePic'])
        except Exception as e:
            print(f"Error decoding image for {name}: {e}")
            profile_pic_blob = None
    else:
        profile_pic_blob = None

    # Insert into DB
    cursor.execute('''
        INSERT INTO students (name, email, contact, profile_pic, profile_pic_type)
        VALUES (?, ?, ?, ?, ?)
    ''', (name, email, contact, profile_pic_blob, profile_pic_type))

conn.commit()
conn.close()
print(" All student records saved in pimeet.db")
```
